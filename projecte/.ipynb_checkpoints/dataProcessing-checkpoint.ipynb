Check all the species that are already defined in 'specie' and add the extra information of the table 'aux'.

```python
for i, row in especie.iterrows():
        
    nombre = especie.iloc[i, 0]
    
    if nombre in aux['scientificName'].values:
        
        # Get the row in aux table
        a = aux[aux['scientificName'] == nombre].iloc[0]
        
        #Check if it has a synonym and add it
        if a['scientificName.1'] != nombre:
            especie.loc[i,['nombre2']] = a['scientificName.1']
            
        # Check if has taxonConceptId
        if not pd.isnull(a['taxonConceptID']):
            especie.loc[i,['taxonConceptID']] = a['taxonConceptID']
             
        # Check if has taxonConceptId
        if not pd.isnull(a['subgenus']):
            especie.loc[i,['subgenus']] = a['subgenus']
            
        # Check if has taxon rank
        if not pd.isnull(a['taxonRank']):
            especie.loc[i,['taxonRank']] = a['taxonRank']

    elif nombre in aux['scientificName.1'].values:
        
        # Get the row in aux table
        a = aux[aux['scientificName.1'] == nombre].iloc[0]
        
        #Check if it has a synonym and add it
        if a['scientificName'] != nombre:
            especie.loc[i,['nombre2']] = a['scientificName']
            
        # Check if has taxonConceptId
        if not pd.isnull(a['taxonConceptID']):
            especie.loc[i,['taxonConceptID']] = a['taxonConceptID']
             
        # Check if has taxonConceptId
        if not pd.isnull(a['subgenus']):
            especie.loc[i,['subgenus']] = a['subgenus']
            
        # Check if has taxon rank
        if not pd.isnull(a['taxonRank']):
            especie.loc[i,['taxonRank']] = a['taxonRank']              
```

Get the cliamte region for each sighting and add it as an attribute

```python
# Latitud --> paralelo respecto ecuador, SUD --> disminuye NORTE --> augmenta
# Longitud --> paralelo meridiano Greenwich ESTE --> augmenta OESTE --> disminuye

climas = ['Oceanico', 'Continental', 'Mediterraneo', 'Montaña', 'Subtropical', 'Semidesertico']

def obtain_climate_region(latitud, longitud):
    # Islas Baleares
    if latitud >= 38.617812 and latitud <= 40.185923 and longitud >= 0.976299 and longitud <= 4.766569:
        return "Mediterraneo"
    # Costa catalana, valenciana y murciana
    elif latitud >= 38.556630 and latitud <= 42.507920 and longitud >= -1.826060 and longitud <= 3.950784:
        return "Mediterraneo"
    # Málaga, Andalucia y sud de Extremadura
    elif latitud >= 35.795098 and latitud <= 39.017897 and longitud >= -6.967191 and longitud <= -3.397742:
        # Sierra de las Nieves 
        if latitud >= 35.795098 and latitud <= 36.818604 and longitud >= -5.463332 and longitud <= -4.971509:
            return 'Montaña'
        # Parque Natural Montes de Málaga 
        elif latitud >= 36.771231 and latitud <= 36.869366 and longitud >= -4.625016 and longitud <= -4.240628:
            return 'Montaña'
        return "Mediterraneo"
    # Región central de la península
    elif latitud >= 37.819820 and latitud <= 42.833947 and longitud >= -6.235231 and longitud <= -1.723790:
        # Sierra de Francia/Gata
        if latitud >= 40.220643 and latitud <= 40.572925 and longitud >= -6.880498 and longitud <= -5.909284:
            return 'Montaña'
        # Sierra de Gredos/Tamajón
        elif latitud >= 40.380539 and latitud <= 41.249557 and longitud >= -5.044052 and longitud <= -3.175033:
            return 'Montaña'
        # Sierra de Tirón/Guadarrama
        elif latitud >= 41.907106 and latitud <= 42.301143 and longitud >= -3.386683 and longitud <= -2.312574:
            return 'Montaña'
        # Serrania de Cuenca
        elif latitud >= 41.907106 and latitud <= 42.301143 and longitud >= -3.386683 and longitud <= -2.312574:
            return 'Montaña'
        # Sierras de Cazorla, Segura y Las Villas 
        elif latitud >= 38.523793 and latitud <= 42.301143 and longitud >= -3.386683 and longitud <= -1.861799:
            return 'Montaña'
        # Riera del Ebro por Aragón
        elif latitud >= 39.649699 and latitud <= 40.728562 and longitud >= -1.457609 and longitud <= -1.462747:
            return 'Semidesertico'
        return "Continental"
    # Galicia
    elif latitud >= 41.788945 and latitud <= 43.691119 and longitud >= -9.565874 and longitud <= -6.128470:
        # Cordillera Cantabrica
        if latitud >= 42.131723 and latitud <= 42.859869 and longitud >= -7.342574 and longitud <= -6.214440:
            return 'Montaña'
        return "Oceanico"
    # Costa Mar Cantabrico
    elif latitud >= 42.621838 and latitud <= 43.754703 and longitud >= -6.128470 and longitud <= -1.812319:
        # Cordillera Cantabrica
        if latitud >= 42.877835 and latitud <= 43.118087 and longitud >= -6.585371 and longitud <= -4.626571:
            return 'Montaña'
        return "Oceanico" 
    # Pre Pirineos
    elif latitud >= 41.836362 and latitud <= 42.995785 and longitud >= -1.724987 and longitud <= 1.248383:
        return "Oceanico" 
    # Pirineos
    elif latitud >= 42.171698 and latitud <= 43.192100 and longitud >= -1.300549 and longitud <= 2.156471:
        return "Montaña" 
    # Costa murciana y Alicante
    elif latitud >= 36.946948 and latitud <= 38.569549 and longitud >= -3.267356 and longitud <= 0.323573:
        return "Semidesértico" 
    # Islas Canarias
    elif latitud >= 27.553322 and latitud <= 29.478362 and longitud >= -19.031080 and longitud <= -13.064945:
        # Teide
        if latitud >= 28.140816 and latitud <= 28.339219 and longitud >= -16.699055 and longitud <= -16.474196:
            return 'Montaña'
        return "Subtropical" 
```

On the other hand, "scientificName" and "scientificName.1" have also the same information. The only difference is that in some cases "scientificName.1" has the synonymous name for the species in "scientificName". 

We already have this information in the table "especie", therefor we will just keep one column with the specie PK of relation "especie". 

```python
def getname(x,y):
    if x in especie['nombre'].values:
        return x
    elif y in especie['nombre'].values:
        return y
    elif x in especie['nombre2'].values:
        return especie['nombre'].where(especie['nombre2'] == x)
    elif y in especie['nombre2'].values:
        return especie['nombre'].where(especie['nombre2'] == y)
    else:
        print(x,y)
        
avistamiento['especie'] = [ getname(x,y) for x,y in zip(avistamiento['scientificName'], avistamiento['scientificName'])]
```

We will classicate the events as if they are related with temperature, either precipitation, either wind.

```python
def isTemperatura(x):
    return '°C' in x.lower() or 'temperatura' in x.lower() or 'calor' in x.lower() or 'frio' in x.lower()

def isViento(x):
    return 'km/h' in x.lower() or 'viento' in x.lower() or 'm/h' in x.lower() or 'tornado' in x.lower() or 'huracan' in x.lower()

def isLluvia(x):
    return 'l/m' in x.lower() or 'lluvias' in x.lower() or 'tormenta' in x.lower() or 'tempestad' in x.lower()
```

```python
eventos['temperatura'] = [ 1 if isTemperatura(x) else 0 for x in eventos['Description']]
eventos['viento'] = [ 1 if isViento(x) else 0 for x in eventos['Description']]
eventos['lluvia'] = [ 1 if isLluvia(x) else 0 for x in eventos['Description']]
```

```python
eventos = eventos.rename(columns={"Day" : "day", "Month" : "month", "Year" : "year", "Description" : "descripcion"}, errors="raise")
```
